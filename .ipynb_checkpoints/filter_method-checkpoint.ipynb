```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import chi2, f_classif, mutual_info_classif
```

```python
url = "https://raw.githubusercontent.com/rikhuijzer/heart-disease-dataset/main/heart-disease-dataset.csv"
df = pd.read_csv(url)
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
missing_percent = X.isnull().mean() * 100
```

```python
X_filled = X.fillna(X.median())
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_filled)
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_filled)
```

```python
chi_scores, _ = chi2(X_scaled, y)
chi_rank = pd.Series(chi_scores, index=X.columns)
```

```python
f_scores, _ = f_classif(X_filled, y)
f_rank = pd.Series(f_scores, index=X.columns)
```

```python
mi_scores = mutual_info_classif(X_filled, y)
mi_rank = pd.Series(mi_scores, index=X.columns)
```

```python
results = pd.DataFrame({
    "Missing %": missing_percent,
    "Chi-Square": chi_rank,
    "F-score": f_rank,
    "Information Gain": mi_rank
})
```

```python
results_sorted = results.sort_values(by="Information Gain", ascending=False)
```

```python
print(results_sorted)
```
